```python
from pathlib import PurePosixPath
from sys import prefix
# Reemplazar con tu API Key de OpenWeather
API_KEY = 'f82e8a17893523fc239a2b36770edbf3'

# Lista de ciudades
ciudades = ['Orlando', 'Sarasota', 'Tampa', 'Jacksonville', 'Miami','Fort Lauderdale','Daytona Beach']

# Lista para almacenar los datos de temperatura
datos_meteorologicos = []

for ciudad in ciudades:
    url = f'http://api.openweathermap.org/data/2.5/weather?q={ciudad},&appid={API_KEY}&units=metric'
    respuesta = requests.get(url)

    if respuesta.status_code == 200:
        datos = respuesta.json()
        if 'main' in datos:
            Fecha = datos['dt']
            temperatura_actual = datos['main']['temp']
            humedad_actual = datos['main']['humidity']
            presion_atmosferica = datos['main']['pressure']
            presion_atmosferica_nivel_mar = datos['main']['sea_level']
            presion_atmosferica_nivel_suelo = datos['main']['grnd_level']
            velocidad_viento = datos['wind']['speed']
            direccion_viento = datos['wind']['deg']
            datos_meteorologicos.append({'Fecha': Fecha,'Ciudad': ciudad, 'Temperatura': temperatura_actual, 'Humedad': humedad_actual,'Presion atmosferica': presion_atmosferica,'Presion atmosferica sobre el nivel del mar':presion_atmosferica_nivel_mar,'Presion atmosferica sobre el nivel del suelo':presion_atmosferica_nivel_suelo,'Velocidad del viento':velocidad_viento,'Direccion del viento':direccion_viento})


        else:
            print(f'Error: No se encontraron los datos de {ciudad}.')
    else:
        print(f'Error: No se pudo obtener los datos del clima para {ciudad}. Código de estado: {respuesta.status_code}')

# Crear un DataFrame de Pandas con los datos de temperatura
df = pd.DataFrame(datos_meteorologicos)

df
```
