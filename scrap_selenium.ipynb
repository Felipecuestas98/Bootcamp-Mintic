```python
# !apt-get update
# !apt install -y chromium-chromedriver
# !pip install selenium

# Importar bibliotecas necesarias
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# Configuración de Selenium para Google Colab
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ejecuta el navegador en modo headless (sin interfaz gráfica)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")


# Iniciar el navegador
# Replacing your service definition with the following, which automatically detects the chromedriver:
driver = webdriver.Chrome(options=chrome_options)


# URL objetivo
url = "https://quotes.toscrape.com/"  # Ejemplo: un sitio de citas públicas para scraping

# Navegar a la URL
driver.get(url)

# Esperar un momento para que la página cargue completamente
time.sleep(2)

# Extraer datos
quotes = driver.find_elements(By.CLASS_NAME, "quote")
data = []

for quote in quotes:
    text = quote.find_element(By.CLASS_NAME, "text").text
    author = quote.find_element(By.CLASS_NAME, "author").text
    data.append({"quote": text, "author": author})

# Mostrar los datos extraídos
for item in data:
    print(f"'{item['quote']}' - {item['author']}")

# Cerrar el navegador
driver.quit()
```

```python
# guarda el resultado anterior en un data frame
import pandas as pd

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
display(df)

# Save DF to csv)
#df.to_csv('quotes.csv', index=False, encoding='utf-8')
```
